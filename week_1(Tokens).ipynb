```python
# removing punctuations
import re
monty = text6.tokens
monty_np = []
for word in monty:
    if re.search('\w', word):
        monty_np.append(word.lower())
len(monty_np)   
```

```python
#compute FreqDist
monty_FreqDist = nltk.FreqDist(monty_np)
most_common = monty_FreqDist.most_common()
most_common
for each_word in most_common:
    print(each_word)
```

```python
#compute MLEProbDist
monty_ProbDist = nltk.MLEProbDist(monty_FreqDist)
monty_ProbDist
```

```python
sample = [(x,monty_ProbDist.prob(x)) for x in monty_ProbDist.samples()]
sample[0:5]
```

# simple plot using pylab

```python
import pylab
def make_plot(x, y, xlab, ylab, scale):
    fig = pylab.figure()
    pylab.plot(x, y)
    pylab.xlabel(xlab)
    pylab.ylabel(ylab)
    if scale == "log":
        pylab.xscale('log')
        pylab.yscale('log')
    pylab.draw()
    pylab.pause(1)
    input("Press any key to continue")
    pylab.close(fig)
olympic_years = [1996, 2000, 2004, 2008, 2012]
winning_times = [500, 400, 300, 200, 100]
make_plot(olympic_years, winning_times, "olympic year", "winning times", "linear")
make_plot(olympic_years, winning_times, "olympic year", "winning times", "log")    
```
