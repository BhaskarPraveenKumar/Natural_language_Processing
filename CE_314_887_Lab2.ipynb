```python
import nltk
from nltk.book import *
from nltk import re 
md_tokens=text1.tokens
md_tokens_nopunct=[word.lower() for word in md_tokens if re.search("\w",word)]
pmld_md=md_tokens_nopunct.count("whale")/len(md_tokens_nopunct)
print (pmld_md)
```

```python
from nltk import word_tokenize
mysent="The cat sat on the mat."
mysent_tokens_nopunct=[word.lower() for word in word_tokenize(mysent) if re.search("\w",word)]
from nltk import FreqDist
mysent_fdist=FreqDist(mysent_tokens_nopunct)
print (mysent_fdist)
print (mysent_fdist.most_common(3))
from nltk import MLEProbDist
mysent_pdist=MLEProbDist(mysent_fdist)
print (mysent_pdist)
```

```python
for x in mysent_pdist.samples():
    print (x,mysent_pdist.prob(x))
```
